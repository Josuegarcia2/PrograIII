```python
print("Sumar dos números: ")
num1 = input("Num 1: ")
num2 = input("Num 2: ")
respuesta = float(num1) + float(num2)

print(type(num1), type(num2), type(respuesta))

print("Respuesta: ", respuesta)

respuesta = "Ssuma completa"
print(respuesta, type(respuesta))
```

```python
print("Restar dos números: ")
num1 = input("Num 1: ")
num2 = input("Num 2: ")
respuesta = float(num1) - float(num2)

print(type(num1), type(num2), type(respuesta))

print("Respuesta: ", respuesta)

respuesta = "Resta completa"
print(respuesta, type(respuesta))
```

```python
print("Multiplicar dos números: ")
num1 = input("Num 1: ")
num2 = input("Num 2: ")
respuesta = float(num1) * float(num2)

print(type(num1), type(num2), type(respuesta))

print("Respuesta: ", respuesta)

respuesta = "Multiplicación completa"
print(respuesta, type(respuesta))
```
